**Reading Data**


```python
#reading Header's
#print('\t\tHeaders\n ', df.columns)


#read each column
print('\n\n:this is Name column from 0 to 5:\n\n')
print(df['Name'][0:5]) #reading one column only
print(df[['Name', 'Type 1', 'HP']] [0:5]) #more than one column


#Reading Each Row
print(df.iloc[1])
print(df.iloc[1:4])


#Read Specific Location
print(df.iloc[2,1])  #row 2 and column 1
df.loc[df['Type 1'] == 'Fire']


df.loc[df['Type 1'] == 'Fire']
df.loc[df['HP'] > 50]
df.loc[df['HP'] > 50].sort_values('HP', ascending=True) #sort in asceing also
```

**Sorting and Describing the Data**

```python
df.describe()
```

```python
df.sort_values('Name', ascending=False)
df.sort_values(['Type 1', 'HP'], ascending=[1,0])
```

**Making Changes to the Data**

```python
df['Total'] = df.loc[:, ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].sum(axis=1)
df.head(5)

#ADDING A NEW COLUMN AT THE LAST HERE
```

```python
#drop a specific column
df.drop(columns=['Total'])
```

```python
#another way for adding column
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
df.head(5)
```

```python
#now change the position of any column and add it between existing columns
cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

# **Save Data**

```python
df.to_csv('updatedDataCSV.csv', index=False)
df.to_excel('updatedDataXL.xlsx', index=False)
df.to_csv('updatedDataTXT.txt', index=False, sep='\t')
```
